```python
# defect_pipeline_colab.py
# Run in Google Colab / Jupyter to show plots inline.
from __future__ import annotations
import os
import io
import sys
import argparse
import urllib.request
import pickle
from typing import Dict, Tuple, Optional

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report, roc_curve, auc,
    precision_recall_curve, average_precision_score
)

# Optional libs
try:
    from xgboost import XGBClassifier  # type: ignore
    HAS_XGB = True
except Exception:
    HAS_XGB = False

try:
    from imblearn.over_sampling import SMOTE  # type: ignore
    HAS_SMOTE = True
except Exception:
    HAS_SMOTE = False

# Visualization
try:
    import matplotlib.pyplot as plt  # type: ignore
    import seaborn as sns  # type: ignore
    HAS_PLOTTING = True
    sns.set(style='whitegrid')
except Exception:
    HAS_PLOTTING = False

# Detect notebook environment (Colab/Jupyter)
def in_notebook() -> bool:
    try:
        get_ip = get_ipython  # type: ignore
        return True
    except Exception:
        return False

# If in notebook, ensure inline backend (Colab/Jupyter)
if in_notebook() and HAS_PLOTTING:
    try:
        # For Colab / Jupyter this is generally enough; users can run %matplotlib inline in a cell.
        import matplotlib as mpl  # type: ignore
        mpl.rcParams['figure.dpi'] = 110
    except Exception:
        pass

# Defaults
DATASETS = {
    'jm1': 'https://raw.githubusercontent.com/ApoorvaKrisna/NASA-promise-dataset-repository/main/jm1.csv',
    'kc1': 'https://raw.githubusercontent.com/ApoorvaKrisna/NASA-promise-dataset-repository/main/kc1.csv',
    'cm1': 'https://raw.githubusercontent.com/ApoorvaKrisna/NASA-promise-dataset-repository/main/cm1.csv',
    'kc2': 'https://raw.githubusercontent.com/ApoorvaKrisna/NASA-promise-dataset-repository/main/kc2.csv',
    'pc1': 'https://raw.githubusercontent.com/ApoorvaKrisna/NASA-promise-dataset-repository/main/pc1.csv',
}

OUTPUT_DIR = 'outputs'
os.makedirs(OUTPUT_DIR, exist_ok=True)


# -------------------- Utility functions --------------------

def download_csv(url: str) -> pd.DataFrame:
    """Download CSV from URL and return DataFrame."""
    print(f"[i] Downloading: {url}")
    resp = urllib.request.urlopen(url)
    data = resp.read()
    return pd.read_csv(io.BytesIO(data))


def detect_label_column(df: pd.DataFrame) -> Optional[str]:
    """Try to detect the defect label column. Common names: 'defects','bug','defective','label','defect'"""
    candidates = ['defects', 'defect', 'bug', 'bugs', 'defective', 'label', 'class']
    lower_cols = {c.lower(): c for c in df.columns}
    for cand in candidates:
        if cand in lower_cols:
            return lower_cols[cand]
    # fallback: if last column looks like a label
    last_col = df.columns[-1]
    if pd.api.types.is_numeric_dtype(df[last_col]):
        uniques = pd.Series(df[last_col].dropna().unique())
        # binary 0/1 or small number of unique ints
        if set(uniques.tolist()) <= {0, 1} or len(uniques) <= 10:
            return last_col
    return None


def prepare_features_labels(df: pd.DataFrame, label_col: Optional[str] = None) -> Tuple[pd.DataFrame, pd.Series]:
    """Return (X, y) where y is binary (0/1) and X contains numeric features only with imputed values."""
    if label_col is None:
        label_col = detect_label_column(df)
    if label_col is None:
        raise ValueError("Could not detect label column automatically. Provide label_col explicitly.")

    print(f"[i] Using label column: {label_col}")
    y = df[label_col].copy()

    # Convert to binary 0/1
    if pd.api.types.is_numeric_dtype(y):
        y = (y > 0).astype(int)
    else:
        y = y.astype(str).str.lower().map({'true':1,'false':0,'yes':1,'no':0}).fillna(0).astype(int)

    X = df.drop(columns=[label_col])
    # keep numeric features only
    numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
    if not numeric_cols:
        raise ValueError("No numeric features detected after dropping the label column.")
    X = X[numeric_cols]

    # drop constant columns
    nunique = X.nunique(dropna=True)
    const_cols = nunique[nunique <= 1].index.tolist()
    if const_cols:
        print(f"[i] Dropping constant columns: {const_cols}")
        X = X.drop(columns=const_cols)

    # impute missing values (median)
    imputer = SimpleImputer(strategy='median')
    X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns, index=X.index)
    return X_imputed, y


# -------------------- Visualization helper functions --------------------

def show_dataset_once(df: pd.DataFrame, n: int = 5):
    """
    Print a concise view of the dataset: columns and first `n` rows.
    This is intended to be called once per dataset to verify contents.
    """
    print("\n=== Dataset preview (columns) ===")
    for i, col in enumerate(df.columns):
        print(f"{i+1}. {col}")
    print("\n=== First {} rows ===".format(n))
    # If running in a notebook, display() will be helpful; else we print.
    try:
        from IPython.display import display  # type: ignore
        display(df.head(n))
    except Exception:
        print(df.head(n).to_string())


def plot_confusion_matrix(cm, model_name: str, dataset_name: str, show: bool = True, save: bool = False):
    if not HAS_PLOTTING:
        return
    plt.figure(figsize=(4.5, 4))
    sns.heatmap(np.array(cm), annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {model_name} ({dataset_name})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    if save:
        out = os.path.join(OUTPUT_DIR, f'cm_{dataset_name}_{model_name}.png')
        plt.savefig(out)
        print(f"[i] Saved confusion matrix to {out}")
    if show:
        plt.show()
    plt.close()


def plot_roc_curve(y_test, y_proba, model_name: str, dataset_name: str, show: bool = True, save: bool = False):
    if not HAS_PLOTTING:
        return
    if y_proba is None or len(np.unique(y_test)) < 2:
        return
    try:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(6, 4))
        plt.plot(fpr, tpr, lw=2, label=f'AUC = {roc_auc:.3f}')
        plt.plot([0, 1], [0, 1], linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve: {model_name} ({dataset_name})')
        plt.legend(loc='lower right')
        plt.tight_layout()
        if save:
            out = os.path.join(OUTPUT_DIR, f'roc_{dataset_name}_{model_name}.png')
            plt.savefig(out)
            print(f"[i] Saved ROC curve to {out}")
        if show:
            plt.show()
        plt.close()
    except Exception as e:
        print(f"[!] Could not plot ROC for {model_name}: {e}")


def plot_precision_recall(y_test, y_proba, model_name: str, dataset_name: str, show: bool = True, save: bool = False):
    """
    Plot Precision-Recall curve and show/save it. Also compute average precision.
    """
    if not HAS_PLOTTING:
        return
    if y_proba is None or len(np.unique(y_test)) < 2:
        return
    try:
        prec, rec, _ = precision_recall_curve(y_test, y_proba)
        ap = average_precision_score(y_test, y_proba)
        plt.figure(figsize=(6, 4))
        plt.plot(rec, prec, lw=2, label=f'AP = {ap:.3f}')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title(f'Precision-Recall: {model_name} ({dataset_name})')
        plt.legend(loc='lower left')
        plt.tight_layout()
        if save:
            out = os.path.join(OUTPUT_DIR, f'pr_{dataset_name}_{model_name}.png')
            plt.savefig(out)
            print(f"[i] Saved Precision-Recall curve to {out}")
        if show:
            plt.show()
        plt.close()
    except Exception as e:
        print(f"[!] Could not plot PR for {model_name}: {e}")


def plot_metric_bar(results: Dict[str, dict], dataset_name: str, show: bool = True, save: bool = False):
    if not HAS_PLOTTING:
        return
    df = pd.DataFrame([{'model': k, 'f1': v['f1'], 'accuracy': v['accuracy']} for k, v in results.items()])
    plt.figure(figsize=(8, 4))
    sns.barplot(data=df, x='model', y='f1')
    plt.title(f'Model F1 Scores ({dataset_name})')
    plt.tight_layout()
    if save:
        out = os.path.join(OUTPUT_DIR, f'f1_{dataset_name}.png')
        plt.savefig(out)
        print(f"[i] Saved f1 bar chart to {out}")
    if show:
        plt.show()
    plt.close()


def compare_models_table(results: Dict[str, dict], dataset_name: str, show: bool = True, save: bool = False):
    """
    Create a comparison DataFrame for the models, save CSV, print, and draw a grouped bar chart and heatmap.
    """
    rows = []
    for m, v in results.items():
        rows.append({
            'model': m,
            'accuracy': v.get('accuracy'),
            'precision': v.get('precision'),
            'recall': v.get('recall'),
            'f1': v.get('f1'),
            'roc_auc': v.get('roc_auc')
        })
    cmp_df = pd.DataFrame(rows).set_index('model')
    # save CSV optionally
    if save:
        csv_out = os.path.join(OUTPUT_DIR, f'comparison_{dataset_name}.csv')
        cmp_df.to_csv(csv_out)
        print(f"[i] Saved comparison table to {csv_out}")

    # print nicely
    print("\n=== Model comparison ===")
    print(cmp_df.round(4).to_string())

    # plot bars & heatmap
    if HAS_PLOTTING:
        cmp_plot = cmp_df[['accuracy','precision','recall','f1']].copy()
        cmp_plot = cmp_plot.reset_index().melt(id_vars='model', var_name='metric', value_name='value')
        plt.figure(figsize=(10,5))
        sns.barplot(data=cmp_plot, x='model', y='value', hue='metric')
        plt.title(f'Model comparison ({dataset_name})')
        plt.legend(loc='upper right')
        plt.tight_layout()
        if save:
            bar_out = os.path.join(OUTPUT_DIR, f'comparison_bar_{dataset_name}.png')
            plt.savefig(bar_out)
            print(f"[i] Saved comparison bar chart to {bar_out}")
        if show:
            plt.show()
        plt.close()

        plt.figure(figsize=(6,4))
        sns.heatmap(cmp_df.fillna(0), annot=True, fmt='.3f', cmap='vlag')
        plt.title('Comparison heatmap')
        plt.tight_layout()
        if save:
            hm_out = os.path.join(OUTPUT_DIR, f'comparison_heatmap_{dataset_name}.png')
            plt.savefig(hm_out)
            print(f"[i] Saved comparison heatmap to {hm_out}")
        if show:
            plt.show()
        plt.close()
    return cmp_df


# -------------------- Modeling/Evaluation functions --------------------

def get_models(random_state: int = 42) -> Dict[str, object]:
    models = {
        'LogisticRegression': LogisticRegression(max_iter=1000, class_weight='balanced', random_state=random_state),
        'DecisionTree': DecisionTreeClassifier(random_state=random_state),
        'RandomForest': RandomForestClassifier(n_estimators=100, random_state=random_state),
        'SVM': SVC(probability=True, class_weight='balanced', random_state=random_state)
    }
    if HAS_XGB:
        models['XGBoost'] = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=random_state)
    return models


def evaluate_model(clf, X_train, X_test, y_train, y_test) -> Tuple[dict, np.ndarray, Optional[np.ndarray]]:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_proba = None
    if hasattr(clf, 'predict_proba'):
        try:
            y_proba = clf.predict_proba(X_test)[:, 1]
        except Exception:
            y_proba = None
    else:
        try:
            y_proba = clf.decision_function(X_test)
            y_proba = np.array(y_proba)
        except Exception:
            y_proba = None

    metrics = {
        'accuracy': float(accuracy_score(y_test, y_pred)),
        'precision': float(precision_score(y_test, y_pred, zero_division=0)),
        'recall': float(recall_score(y_test, y_pred, zero_division=0)),
        'f1': float(f1_score(y_test, y_pred, zero_division=0)),
    }

    if y_proba is not None and len(np.unique(y_test)) > 1:
        try:
            auc_val = float(roc_auc_score(y_test, y_proba))
            metrics['roc_auc'] = auc_val
        except Exception:
            metrics['roc_auc'] = None
    else:
        metrics['roc_auc'] = None

    cm = confusion_matrix(y_test, y_pred)
    metrics['confusion_matrix'] = cm.tolist()
    return metrics, y_pred, y_proba


def save_classification_report(y_true, y_pred, model_name: str, dataset_name: str, save: bool = False):
    report = classification_report(y_true, y_pred, zero_division=0)
    if save:
        out = os.path.join(OUTPUT_DIR, f'classification_report_{dataset_name}_{model_name}.txt')
        with open(out, 'w') as f:
            f.write(report)
        print(f"[i] Saved classification report to {out}")
    return report


def save_model_pickle(model, model_name: str, dataset_name: str, save: bool = False):
    if save:
        out = os.path.join(OUTPUT_DIR, f'model_{dataset_name}_{model_name}.pkl')
        with open(out, 'wb') as f:
            pickle.dump(model, f)
        print(f"[i] Saved model pickle to {out}")


# -------------------- Run for single dataset --------------------

def run_for_dataset(name: str, url: str, use_smote: bool = True, test_size: float = 0.2,
                    random_state: int = 42, save_artifacts: bool = False, show_plots: Optional[bool] = None,
                    label_col: Optional[str] = None) -> Dict[str, dict]:
    """Runs pipeline for a single dataset. If show_plots is None, notebook mode => True else False."""
    if show_plots is None:
        show_plots = in_notebook()

    df = download_csv(url)
    print(f"[i] {name} raw shape: {df.shape}")

    # ---- show dataset once ----
    show_dataset_once(df, n=5)

    X, y = prepare_features_labels(df, label_col=label_col)
    print(f"[i] Features shape: {X.shape}  Labels distribution:\n{y.value_counts().to_dict()}")

    # train-test split (stratified)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, stratify=y, random_state=random_state
    )

    # scale: fit on training only
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # SMOTE on scaled training set
    if use_smote and HAS_SMOTE:
        sm = SMOTE(random_state=random_state)
        X_train_res, y_train_res = sm.fit_resample(X_train_scaled, y_train)
        print(f"[i] After SMOTE, train distribution: {pd.Series(y_train_res).value_counts().to_dict()}")
    else:
        X_train_res, y_train_res = X_train_scaled, y_train
        if use_smote and not HAS_SMOTE:
            print("[!] imblearn.SMOTE not installed; continuing without SMOTE")

    results: Dict[str, dict] = {}
    models = get_models(random_state=random_state)

    for mname, model in models.items():
        print(f"\n[i] Training {mname} on {name}")
        try:
            metrics, y_pred, y_proba = evaluate_model(model, X_train_res, X_test_scaled, y_train_res, y_test)
            results[mname] = metrics
            print(f"[i] Metrics: {metrics}")

            # Save / show artifacts
            save_classification_report(y_test, y_pred, mname, name, save=save_artifacts)
            save_model_pickle(model, mname, name, save=save_artifacts)

            # plot & show inline in Colab/Jupyter (and optionally save)
            plot_confusion_matrix(metrics['confusion_matrix'], mname, name, show=show_plots, save=save_artifacts)
            plot_roc_curve(y_test, y_proba, mname, name, show=show_plots, save=save_artifacts)
            plot_precision_recall(y_test, y_proba, mname, name, show=show_plots, save=save_artifacts)

        except Exception as e:
            print(f"[!] Error training/evaluating {mname} on {name}: {e}")

    # Save summary CSV for this dataset (optional)
    rows = []
    for k, v in results.items():
        rows.append({
            'dataset': name,
            'model': k,
            'accuracy': v.get('accuracy'),
            'precision': v.get('precision'),
            'recall': v.get('recall'),
            'f1': v.get('f1'),
            'roc_auc': v.get('roc_auc')
        })
    if save_artifacts:
        out_csv = os.path.join(OUTPUT_DIR, f'results_{name}.csv')
        pd.DataFrame(rows).to_csv(out_csv, index=False)
        print(f"[i] Saved summary to {out_csv}")

    # aggregated visual + table (show inline)
    cmp_df = compare_models_table(results, name, show=show_plots, save=save_artifacts)

    return results


# -------------------- CLI / Main --------------------

def parse_args():
    parser = argparse.ArgumentParser(description="Defect prediction pipeline (Colab/Jupyter inline visuals)")
    parser.add_argument('--datasets', nargs='*', default=None,
                        help='Optional list of dataset_url:name pairs like "https://.../file.csv:mydata". If omitted, the default PROMISE list will be used.')
    parser.add_argument('--use-smote', action='store_true', default=True, help='Apply SMOTE on training set (requires imbalanced-learn).')
    parser.add_argument('--test-size', type=float, default=0.2, help='Test set fraction.')
    parser.add_argument('--random-state', type=int, default=42, help='Random seed.')
    parser.add_argument('--save-artifacts', action='store_true', help='Save images/CSVs to outputs/ directory in addition to showing inline.')
    parser.add_argument('--show-plots', action='store_true', help='Force-show plots (useful if running as a script outside notebook).')
    return parser.parse_args([])

def main(argv=None):
    if argv is None:
        args = parse_args()
    else:
        # if called programmatically, accept list
        args = parse_args() if isinstance(argv, list) and len(argv)==0 else parse_args()

    # build dataset list
    if args.datasets:
        ds_map = {}
        for item in args.datasets:
            # expect either "url" or "url:name"
            if ':' in item:
                url, name = item.split(':', 1)
            else:
                url = item
                name = os.path.splitext(os.path.basename(url))[0]
            ds_map[name] = url
    else:
        ds_map = DATASETS.copy()

    # If running in notebook, default to showing plots
    notebook = in_notebook()
    show_plots = True if notebook else args.show_plots

    all_results = {}
    for name, url in ds_map.items():
        label_col_for_current_dataset = None
        if name == 'kc2':
            # Manually setting 'defective' as the label column for kc2
            label_col_for_current_dataset = 'defective'
            print(f"[i] Manually setting label column for {name} to '{label_col_for_current_dataset}'")
        try:
            res = run_for_dataset(name, url,
                                  use_smote=args.use_smote,
                                  test_size=args.test_size,
                                  random_state=args.random_state,
                                  save_artifacts=args.save_artifacts,
                                  show_plots=show_plots,
                                  label_col=label_col_for_current_dataset) # Pass the label_col
            all_results[name] = res
        except Exception as e:
            print(f"[!] Error running dataset {name}: {e}")

    # Optionally save combined CSV
    combined = []
    for d, rset in all_results.items():
        for m, mets in rset.items():
            combined.append({
                'dataset': d, 'model': m,
                'accuracy': mets.get('accuracy'),
                'precision': mets.get('precision'),
                'recall': mets.get('recall'),
                'f1': mets.get('f1'),
                'roc_auc': mets.get('roc_auc')
            })
    if args.save_artifacts and combined:
        combined_csv = os.path.join(OUTPUT_DIR, 'combined_results.csv')
        pd.DataFrame(combined).to_csv(combined_csv, index=False)
        print(f"[i] Saved combined results to {combined_csv}")

    print("[i] Done.")

if __name__ == '__main__':
    # If you paste this file into Colab and run as a cell, call main() directly:
    # main([])
    # If running from command line, simply run: python defect_pipeline_colab.py
    main()
```

```python
# interactive_pyDriller_full_table.py
# Run in Google Colab / Jupyter or plain Python.
# Requires: pip install pydriller tqdm pandas

from pydriller import Repository
from collections import defaultdict
import statistics
import pandas as pd
from tqdm import tqdm
import sys

def get_repo_url_from_user(default: str = "https://github.com/cxxc7/yantra-chatbot") -> str:
    """
    Prompt the user to input a GitHub repository URL (HTTPS). If empty, returns default.
    """
    try:
        # In notebooks this shows a prompt; in scripts too.
        inp = input(f"Enter GitHub repo URL (or press Enter to use default [{default}]): ").strip()
    except Exception:
        # fallback when input can't be used
        inp = default
    return inp if inp else default

def mine_repo_commits(repo_url: str, show_progress: bool = True) -> pd.DataFrame:
    """
    Traverse the repository commits and extract per-commit metrics into a DataFrame.
    """
    author_commit_count = defaultdict(int)
    records = []

    print(f"[i] Mining repository: {repo_url}")
    print("[!] Warning: showing ALL commits in a table may be slow for large repos (thousands of commits).")

    try:
        iterator = Repository(repo_url).traverse_commits()
    except Exception as e:
        print(f"[!] Error creating Repository iterator: {e}")
        raise

    it = iterator if not show_progress else tqdm(iterator, desc="Processing commits", unit="commit")

    for commit in it:
        try:
            total_added = 0
            total_deleted = 0
            file_count = len(commit.modified_files)
            churn_list = []
            complexity_list = []
            methods_list = []

            # History metric: developer experience
            author_prev_commits = author_commit_count[commit.author.email]
            author_commit_count[commit.author.email] += 1

            # File-level analysis
            for m in commit.modified_files:
                added = m.added_lines or 0
                deleted = m.deleted_lines or 0
                total_added += added
                total_deleted += deleted

                churn = added + deleted
                churn_list.append(churn)

                if m.complexity is not None:
                    complexity_list.append(m.complexity)

                if m.methods:
                    methods_list.append(len(m.methods))

            # Aggregate metrics safely
            avg_churn = statistics.mean(churn_list) if churn_list else 0
            avg_complexity = statistics.mean(complexity_list) if complexity_list else 0
            avg_methods = statistics.mean(methods_list) if methods_list else 0

            records.append({
                "commit_hash": commit.hash,
                "author": commit.author.name,
                "author_email": commit.author.email,
                "author_previous_commits": author_prev_commits,
                "date": commit.author_date,
                "files_modified": file_count,
                "total_added": total_added,
                "total_deleted": total_deleted,
                "avg_file_churn": avg_churn,
                "avg_file_complexity": avg_complexity,
                "avg_file_methods": avg_methods,
                "commit_message_length": len(commit.msg),
                "commit_msg": commit.msg[:500]  # keep at most 500 chars for display
            })
        except Exception as e:
            # Don't stop the whole run if a single commit causes an issue
            print(f"[!] Skipping commit due to error: {e}", file=sys.stderr)
            continue

    df = pd.DataFrame(records)
    return df

def show_full_dataframe(df: pd.DataFrame):
    """
    Show the entire DataFrame in a notebook-friendly way if possible.
    Otherwise print the DataFrame fully in the console.
    """
    # Save CSV
    df.to_csv("commit_metrics.csv", index=False)
    print("\nSaved commit metrics to commit_metrics.csv")

    # Try to display nicely in notebook
    try:
        from IPython.display import display
        pd.set_option('display.max_rows', None)       # show all rows
        pd.set_option('display.max_columns', None)    # show all columns
        pd.set_option('display.width', 200)
        display(df)
    except Exception:
        # Fall back to printing all rows to console
        pd.set_option('display.max_rows', None)
        pd.set_option('display.max_columns', None)
        print(df.to_string())

def main():
    repo_url = get_repo_url_from_user()
    df = mine_repo_commits(repo_url, show_progress=True)
    print(f"\n[i] Extracted {len(df)} commits.")
    # show the full table (careful: may be large)
    show_full_dataframe(df)

if __name__ == "__main__":
    main()
```

```python
# commit_risk_evaluator.py
# Run in Colab / Jupyter or plain Python.
# Requires: pandas, numpy, matplotlib, seaborn (for nicer plots)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Optional

sns.set(style='whitegrid')

def load_commit_metrics(path: str = "commit_metrics.csv") -> pd.DataFrame:
    df = pd.read_csv(path, parse_dates=['date'])
    return df

def compute_normalized_scores(df: pd.DataFrame) -> pd.DataFrame:
    """
    Compute normalized sub-scores in [0,1] for several risk factors.
    Returns a copy of df with added columns: churn_score, complexity_score,
    files_modified_score, author_experience_score, msglen_score.
    """
    d = df.copy()

    # Use robust scaling to [0,1] with percentiles (avoid outlier influence)
    def norm_col(col, lower_pct=5, upper_pct=95):
        vals = d[col].fillna(0).astype(float)
        lo = np.percentile(vals, lower_pct)
        hi = np.percentile(vals, upper_pct)
        if hi - lo <= 0:
            # fallback scalar
            return (vals - vals.min()) / (vals.max() - vals.min() + 1e-9)
        scaled = (vals - lo) / (hi - lo)
        return np.clip(scaled, 0.0, 1.0)

    # churn: use total_added + total_deleted or avg_file_churn if that exists
    if 'total_added' in d.columns and 'total_deleted' in d.columns:
        d['churn'] = d['total_added'].fillna(0) + d['total_deleted'].fillna(0)
    elif 'avg_file_churn' in d.columns:
        d['churn'] = d['avg_file_churn'].fillna(0)
    else:
        d['churn'] = 0

    d['churn_score'] = norm_col('churn')

    # complexity score: avg_file_complexity if present
    if 'avg_file_complexity' in d.columns:
        d['complexity_score'] = norm_col('avg_file_complexity')
    else:
        d['complexity_score'] = 0

    # files modified
    d['files_modified_score'] = norm_col('files_modified') if 'files_modified' in d.columns else 0

    # author experience: note that less experience -> higher risk, so invert
    if 'author_previous_commits' in d.columns:
        raw = d['author_previous_commits'].fillna(0).astype(float)
        exp_score = norm_col('author_previous_commits')
        # invert: inexperienced => score close to 1
        d['author_experience_score'] = 1.0 - exp_score
    else:
        d['author_experience_score'] = 0

    # commit message length: very short messages sometimes correlate with rushed changes
    if 'commit_message_length' in d.columns:
        d['msglen_score'] = 1.0 - norm_col('commit_message_length')  # shorter -> higher score
    else:
        d['msglen_score'] = 0

    # fill NaNs
    for col in ['churn_score','complexity_score','files_modified_score','author_experience_score','msglen_score']:
        d[col] = d[col].fillna(0.0)

    return d

def combine_risk_score(df: pd.DataFrame, weights: Optional[dict] = None) -> pd.DataFrame:
    """
    Combine sub-scores into a single risk_score via weighted average.
    Default weights chosen to reflect typical importance: churn and files modified are strong signals.
    """
    if weights is None:
        weights = {
            'churn_score': 0.35,
            'complexity_score': 0.20,
            'files_modified_score': 0.25,
            'author_experience_score': 0.10,
            'msglen_score': 0.10
        }
    d = df.copy()
    # ensure keys exist
    for k in weights.keys():
        if k not in d.columns:
            d[k] = 0.0
    # weighted sum
    weighted = np.zeros(len(d))
    total_w = sum(weights.values())
    for k,w in weights.items():
        weighted += d[k].values * (w / total_w)
    d['risk_score'] = np.clip(weighted, 0.0, 1.0)
    return d

def classify_risk(df: pd.DataFrame, thresholds=(0.33, 0.66)) -> pd.DataFrame:
    """
    thresholds: (low_to_med, med_to_high)
    Returns df with 'risk_label' column.
    """
    d = df.copy()
    low_th, high_th = thresholds
    def lab(s):
        if s >= high_th:
            return 'HIGH'
        elif s >= low_th:
            return 'MEDIUM'
        else:
            return 'LOW'
    d['risk_label'] = d['risk_score'].apply(lab)
    return d

def summarize_repo_risk(df: pd.DataFrame) -> dict:
    """
    Produce repo-level summary: counts and percentages of risk labels, top risky commits.
    """
    total = len(df)
    counts = df['risk_label'].value_counts().to_dict()
    pct = {k: (v / total * 100) for k,v in counts.items()}
    top_risky = df.sort_values('risk_score', ascending=False).head(10)[
        ['commit_hash','author','date','risk_score','risk_label','churn','avg_file_complexity','files_modified']
    ].reset_index(drop=True)
    return {
        'total_commits': total,
        'counts': counts,
        'percentages': pct,
        'top_risky_commits': top_risky
    }

def visualize_risk(df: pd.DataFrame, show_top=10):
    """
    Inline visuals: histogram of risk scores and bar of top risky commits.
    """
    # histogram
    plt.figure(figsize=(8,4))
    sns.histplot(df['risk_score'], bins=25, kde=True)
    plt.title('Distribution of Commit Risk Scores')
    plt.xlabel('Risk score (0 low - 1 high)')
    plt.show()

    # pie of risk labels
    plt.figure(figsize=(6,4))
    lab_counts = df['risk_label'].value_counts()
    lab_counts.plot.pie(autopct='%1.1f%%', ylabel='')
    plt.title('Percentage by Risk Label')
    plt.show()

    # top risky commits bar
    top = df.sort_values('risk_score', ascending=False).head(show_top)
    plt.figure(figsize=(10, max(4, show_top*0.4)))
    sns.barplot(x='risk_score', y='commit_hash', data=top, orient='h')
    plt.title(f'Top {show_top} risky commits')
    plt.xlabel('Risk score')
    plt.ylabel('Commit hash')
    plt.show()

def evaluate_commits(path_or_df: Optional[str or pd.DataFrame] = "commit_metrics.csv",
                     weights: Optional[dict] = None,
                     thresholds=(0.33,0.66),
                     show_plots: bool = True) -> dict:
    """
    Main entry point. Returns a dictionary with detailed results and prints summary.
    """
    if isinstance(path_or_df, str):
        df = load_commit_metrics(path_or_df)
    else:
        df = path_or_df.copy()

    df2 = compute_normalized_scores(df)
    df3 = combine_risk_score(df2, weights=weights)
    df4 = classify_risk(df3, thresholds=thresholds)

    summary = summarize_repo_risk(df4)

    # print summary
    print("=== Repo Risk Summary ===")
    print(f"Total commits analyzed: {summary['total_commits']}")
    print("Counts by risk label:", summary['counts'])
    print("Percentages by risk label (approx):")
    for k,v in summary['percentages'].items():
        print(f"  {k}: {v:.2f}%")
    print("\nTop risky commits (head):")
    display_cols = ['commit_hash','author','date','risk_score','risk_label','churn','avg_file_complexity','files_modified']
    print(summary['top_risky_commits'][display_cols].to_string(index=False))

    if show_plots:
        visualize_risk(df4, show_top=10)

    # Return results structure
    results = {
        'detailed_df': df4,
        'summary': summary
    }
    return results

# Example usage:
results = evaluate_commits("commit_metrics.csv", show_plots=True)
detailed_df = results['detailed_df']   # has risk_score and risk_label for each commit
summary = results['summary']
```

```python
# interactive_pyDriller_full_table.py
# Run in Google Colab / Jupyter or plain Python.
# Requires: pip install pydriller tqdm pandas

from pydriller import Repository
from collections import defaultdict
import statistics
import pandas as pd
from tqdm import tqdm
import sys

def get_repo_url_from_user(default: str = "https://github.com/cxxc7/yantra-chatbot") -> str:
    """
    Prompt the user to input a GitHub repository URL (HTTPS). If empty, returns default.
    """
    try:
        # In notebooks this shows a prompt; in scripts too.
        inp = input(f"Enter GitHub repo URL (or press Enter to use default [{default}]): ").strip()
    except Exception:
        # fallback when input can't be used
        inp = default
    return inp if inp else default

def mine_repo_commits(repo_url: str, show_progress: bool = True) -> pd.DataFrame:
    """
    Traverse the repository commits and extract per-commit metrics into a DataFrame.
    """
    author_commit_count = defaultdict(int)
    records = []

    print(f"[i] Mining repository: {repo_url}")
    print("[!] Warning: showing ALL commits in a table may be slow for large repos (thousands of commits).")

    try:
        iterator = Repository(repo_url).traverse_commits()
    except Exception as e:
        print(f"[!] Error creating Repository iterator: {e}")
        raise

    it = iterator if not show_progress else tqdm(iterator, desc="Processing commits", unit="commit")

    for commit in it:
        try:
            total_added = 0
            total_deleted = 0
            file_count = len(commit.modified_files)
            churn_list = []
            complexity_list = []
            methods_list = []

            # History metric: developer experience
            author_prev_commits = author_commit_count[commit.author.email]
            author_commit_count[commit.author.email] += 1

            # File-level analysis
            for m in commit.modified_files:
                added = m.added_lines or 0
                deleted = m.deleted_lines or 0
                total_added += added
                total_deleted += deleted

                churn = added + deleted
                churn_list.append(churn)

                if m.complexity is not None:
                    complexity_list.append(m.complexity)

                if m.methods:
                    methods_list.append(len(m.methods))

            # Aggregate metrics safely
            avg_churn = statistics.mean(churn_list) if churn_list else 0
            avg_complexity = statistics.mean(complexity_list) if complexity_list else 0
            avg_methods = statistics.mean(methods_list) if methods_list else 0

            records.append({
                "commit_hash": commit.hash,
                "author": commit.author.name,
                "author_email": commit.author.email,
                "author_previous_commits": author_prev_commits,
                "date": commit.author_date,
                "files_modified": file_count,
                "total_added": total_added,
                "total_deleted": total_deleted,
                "avg_file_churn": avg_churn,
                "avg_file_complexity": avg_complexity,
                "avg_file_methods": avg_methods,
                "commit_message_length": len(commit.msg),
                "commit_msg": commit.msg[:500]  # keep at most 500 chars for display
            })
        except Exception as e:
            # Don't stop the whole run if a single commit causes an issue
            print(f"[!] Skipping commit due to error: {e}", file=sys.stderr)
            continue

    df = pd.DataFrame(records)
    return df

def show_full_dataframe(df: pd.DataFrame):
    """
    Show the entire DataFrame in a notebook-friendly way if possible.
    Otherwise print the DataFrame fully in the console.
    """
    # Save CSV
    df.to_csv("commit_metrics.csv", index=False)
    print("\nSaved commit metrics to commit_metrics.csv")

    # Try to display nicely in notebook
    try:
        from IPython.display import display
        pd.set_option('display.max_rows', None)       # show all rows
        pd.set_option('display.max_columns', None)    # show all columns
        pd.set_option('display.width', 200)
        display(df)
    except Exception:
        # Fall back to printing all rows to console
        pd.set_option('display.max_rows', None)
        pd.set_option('display.max_columns', None)
        print(df.to_string())

def main():
    repo_url = get_repo_url_from_user()
    df = mine_repo_commits(repo_url, show_progress=True)
    print(f"\n[i] Extracted {len(df)} commits.")
    # show the full table (careful: may be large)
    show_full_dataframe(df)

if __name__ == "__main__":
    main()
```

```python
# commit_risk_evaluator.py
# Run in Colab / Jupyter or plain Python.
# Requires: pandas, numpy, matplotlib, seaborn (for nicer plots)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Optional

sns.set(style='whitegrid')

def load_commit_metrics(path: str = "commit_metrics.csv") -> pd.DataFrame:
    df = pd.read_csv(path, parse_dates=['date'])
    return df

def compute_normalized_scores(df: pd.DataFrame) -> pd.DataFrame:
    """
    Compute normalized sub-scores in [0,1] for several risk factors.
    Returns a copy of df with added columns: churn_score, complexity_score,
    files_modified_score, author_experience_score, msglen_score.
    """
    d = df.copy()

    # Use robust scaling to [0,1] with percentiles (avoid outlier influence)
    def norm_col(col, lower_pct=5, upper_pct=95):
        vals = d[col].fillna(0).astype(float)
        lo = np.percentile(vals, lower_pct)
        hi = np.percentile(vals, upper_pct)
        if hi - lo <= 0:
            # fallback scalar
            return (vals - vals.min()) / (vals.max() - vals.min() + 1e-9)
        scaled = (vals - lo) / (hi - lo)
        return np.clip(scaled, 0.0, 1.0)

    # churn: use total_added + total_deleted or avg_file_churn if that exists
    if 'total_added' in d.columns and 'total_deleted' in d.columns:
        d['churn'] = d['total_added'].fillna(0) + d['total_deleted'].fillna(0)
    elif 'avg_file_churn' in d.columns:
        d['churn'] = d['avg_file_churn'].fillna(0)
    else:
        d['churn'] = 0

    d['churn_score'] = norm_col('churn')

    # complexity score: avg_file_complexity if present
    if 'avg_file_complexity' in d.columns:
        d['complexity_score'] = norm_col('avg_file_complexity')
    else:
        d['complexity_score'] = 0

    # files modified
    d['files_modified_score'] = norm_col('files_modified') if 'files_modified' in d.columns else 0

    # author experience: note that less experience -> higher risk, so invert
    if 'author_previous_commits' in d.columns:
        raw = d['author_previous_commits'].fillna(0).astype(float)
        exp_score = norm_col('author_previous_commits')
        # invert: inexperienced => score close to 1
        d['author_experience_score'] = 1.0 - exp_score
    else:
        d['author_experience_score'] = 0

    # commit message length: very short messages sometimes correlate with rushed changes
    if 'commit_message_length' in d.columns:
        d['msglen_score'] = 1.0 - norm_col('commit_message_length')  # shorter -> higher score
    else:
        d['msglen_score'] = 0

    # fill NaNs
    for col in ['churn_score','complexity_score','files_modified_score','author_experience_score','msglen_score']:
        d[col] = d[col].fillna(0.0)

    return d

def combine_risk_score(df: pd.DataFrame, weights: Optional[dict] = None) -> pd.DataFrame:
    """
    Combine sub-scores into a single risk_score via weighted average.
    Default weights chosen to reflect typical importance: churn and files modified are strong signals.
    """
    if weights is None:
        weights = {
            'churn_score': 0.35,
            'complexity_score': 0.20,
            'files_modified_score': 0.25,
            'author_experience_score': 0.10,
            'msglen_score': 0.10
        }
    d = df.copy()
    # ensure keys exist
    for k in weights.keys():
        if k not in d.columns:
            d[k] = 0.0
    # weighted sum
    weighted = np.zeros(len(d))
    total_w = sum(weights.values())
    for k,w in weights.items():
        weighted += d[k].values * (w / total_w)
    d['risk_score'] = np.clip(weighted, 0.0, 1.0)
    return d

def classify_risk(df: pd.DataFrame, thresholds=(0.33, 0.66)) -> pd.DataFrame:
    """
    thresholds: (low_to_med, med_to_high)
    Returns df with 'risk_label' column.
    """
    d = df.copy()
    low_th, high_th = thresholds
    def lab(s):
        if s >= high_th:
            return 'HIGH'
        elif s >= low_th:
            return 'MEDIUM'
        else:
            return 'LOW'
    d['risk_label'] = d['risk_score'].apply(lab)
    return d

def summarize_repo_risk(df: pd.DataFrame) -> dict:
    """
    Produce repo-level summary: counts and percentages of risk labels, top risky commits.
    """
    total = len(df)
    counts = df['risk_label'].value_counts().to_dict()
    pct = {k: (v / total * 100) for k,v in counts.items()}
    top_risky = df.sort_values('risk_score', ascending=False).head(10)[
        ['commit_hash','author','date','risk_score','risk_label','churn','avg_file_complexity','files_modified']
    ].reset_index(drop=True)
    return {
        'total_commits': total,
        'counts': counts,
        'percentages': pct,
        'top_risky_commits': top_risky
    }

def visualize_risk(df: pd.DataFrame, show_top=10):
    """
    Inline visuals: histogram of risk scores and bar of top risky commits.
    """
    # histogram
    plt.figure(figsize=(8,4))
    sns.histplot(df['risk_score'], bins=25, kde=True)
    plt.title('Distribution of Commit Risk Scores')
    plt.xlabel('Risk score (0 low - 1 high)')
    plt.show()

    # pie of risk labels
    plt.figure(figsize=(6,4))
    lab_counts = df['risk_label'].value_counts()
    lab_counts.plot.pie(autopct='%1.1f%%', ylabel='')
    plt.title('Percentage by Risk Label')
    plt.show()

    # top risky commits bar
    top = df.sort_values('risk_score', ascending=False).head(show_top)
    plt.figure(figsize=(10, max(4, show_top*0.4)))
    sns.barplot(x='risk_score', y='commit_hash', data=top, orient='h')
    plt.title(f'Top {show_top} risky commits')
    plt.xlabel('Risk score')
    plt.ylabel('Commit hash')
    plt.show()

def evaluate_commits(path_or_df: Optional[str or pd.DataFrame] = "commit_metrics.csv",
                     weights: Optional[dict] = None,
                     thresholds=(0.33,0.66),
                     show_plots: bool = True) -> dict:
    """
    Main entry point. Returns a dictionary with detailed results and prints summary.
    """
    if isinstance(path_or_df, str):
        df = load_commit_metrics(path_or_df)
    else:
        df = path_or_df.copy()

    df2 = compute_normalized_scores(df)
    df3 = combine_risk_score(df2, weights=weights)
    df4 = classify_risk(df3, thresholds=thresholds)

    summary = summarize_repo_risk(df4)

    # print summary
    print("=== Repo Risk Summary ===")
    print(f"Total commits analyzed: {summary['total_commits']}")
    print("Counts by risk label:", summary['counts'])
    print("Percentages by risk label (approx):")
    for k,v in summary['percentages'].items():
        print(f"  {k}: {v:.2f}%")
    print("\nTop risky commits (head):")
    display_cols = ['commit_hash','author','date','risk_score','risk_label','churn','avg_file_complexity','files_modified']
    print(summary['top_risky_commits'][display_cols].to_string(index=False))

    if show_plots:
        visualize_risk(df4, show_top=10)

    # Return results structure
    results = {
        'detailed_df': df4,
        'summary': summary
    }
    return results

# Example usage:
results = evaluate_commits("commit_metrics.csv", show_plots=True)
detailed_df = results['detailed_df']   # has risk_score and risk_label for each commit
summary = results['summary']
```
